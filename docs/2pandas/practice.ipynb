```python
import pandas as pd
import numpy as np

# ✅ 正确的数据结构
bars = {
    "time": pd.DataFrame(
        {0: [1704067200000], 1: [1704153600000], 2: [1704240000000]},
        index=["000001.SZ"]
    ),
    "open": pd.DataFrame(
        {0: [12.5], 1: [12.7], 2: [12.6]},
        index=["000001.SZ"]
    ),
    "high": pd.DataFrame(
        {0: [12.8], 1: [12.9], 2: [12.7]},
        index=["000001.SZ"]
    ),
    "low": pd.DataFrame(
        {0: [12.3], 1: [12.5], 2: [12.4]},
        index=["000001.SZ"]
    ),
    "close": pd.DataFrame(
        {0: [12.7], 1: [12.6], 2: [12.5]},
        index=["000001.SZ"]
    ),
    "volume": pd.DataFrame(
        {0: [1000000], 1: [1200000], 2: [1100000]},
        index=["000001.SZ"]
    ),
}

```

```python
bars["time"]
```

```python
symbol = "000001.SZ"
loc = bars["time"].loc[symbol]
```

```python
loc
```

```python
loc = loc.values
```

```python
loc
```

```python
idx = pd.DatetimeIndex(loc.flatten())
idx
```

```python
idx = pd.DatetimeIndex(loc)
```

```python
idx
```

```python
idx = pd.DatetimeIndex(pd.to_datetime(loc, unit='ms'))
idx
```

```python
_OHLCV_FIELDS = ("open", "high", "low", "close", "volume")
_OHLCV_COLUMNS = ("Open", "High", "Low", "Close", "Volume")
data = {}
for f, col in zip(_OHLCV_FIELDS, _OHLCV_COLUMNS):
    data[col] = bars[f].loc[symbol].values.flatten()
    print(f"{col}: {data[col]}")
```

```python
df = pd.DataFrame(data, index=idx)
```

```python
df
```
